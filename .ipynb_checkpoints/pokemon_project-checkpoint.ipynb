# Pokemon Dataset

```python
# importing necessary libraries needed to complete this project
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn import preprocessing
from preprocessing import Labe
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
# making the csv readable using pandas pd.read function
pokemon = pd.read_csv('pokemon.csv')
```

# Data Cleaning
removing NA values from the dataset and adding column names to the dataframe using the actual name of the pokemon from the dataset

```python
pokemon['type2'].fillna(pokemon['type1'], inplace=True)
```

```python
pokemon['percentage_male'].fillna(0, inplace=True)
```

```python
pokemon['height_m'].fillna(pokemon['height_m'].mean(), inplace=True)
```

```python
pokemon['weight_kg'].fillna(pokemon['weight_kg'].mean(), inplace=True)
```

```python
pokemon.isnull().sum()
```

```python
pokemon_T = pokemon.T
pokemon_T.rename(columns=pokemon['name'], inplace=True)
```

```python
pokemon_T
```

```python
# using KNeighbors Classification from sklearn to identify if a pokemon is a legendary pokemon with the given dataset

#instantiating a class of the KNeighbors Classifer 
knn = KNeighborsClassifier()

#identifying the feature data and response data
features = pokemon.drop(columns=['is_legendary'])
response = pokemon['is_legendary']

#instantiating a class of the one hot encoder and label encoder from sklearn, fitting and transforming the data
one_hot = OneHotEncoder()
le = preprocessing.LabelEncoder()

features_encoded = one_hot.fit_transform(features)
response_encoded = le.fit_transform(response)

#split the data into training and testing data, with a test size of 20% and shuffeled the data to make it harder for the model to identify
X_train, X_test, y_train, y_test = train_test_split(features_encoded, response_encoded, test_size=0.20, random_state=42, shuffle=True)


#fit the data to the model
knn.fit(X_train, y_train)
```

```python
# predicted the testing data to print the accuracy, percision, and recall score of the data
y_preds = knn.predict(X_test)

print(knn.__class__.__name__, accuracy_score(y_test, y_preds))
print(knn.__class__.__name__, precision_score(y_test, y_preds))
print(knn.__class__.__name__, recall_score(y_test, y_preds))
```

Model is accurate with high precision, returns many false positives. Due because of the shuffling of data back in the train_test_plit. I could achieve a higher recall and accuracy score, but at the expense of overfitting the data. Which is bad
